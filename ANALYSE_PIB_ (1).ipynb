```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données d'exemple synthétiques (à remplacer par vos données réelles complètes mondiales)
data = {
    'Pays': ['États-Unis', 'Chine', 'Inde', 'Brésil', 'Allemagne', 'France', 'Maroc', 'Turquie', 'Arabie Saoudite'],
    'PIB_2023_Md$': [25721, 17948, 3500, 2100, 4200, 2900, 150, 796, 870],
    'PIB_2024_Md$': [26000, 18700, 3650, 2188, 4300, 2950, 155, 838, 936],
    'Population_millions': [331, 1440, 1393, 213, 83, 67, 37, 85, 35],
    'Croissance_2025_%': [2.1, 4.7, 6.2, 1.8, 0.3, 0.6, 2.9, 2.9, 2.4]  # simple prévisions ex.
}

df = pd.DataFrame(data)

# Calcul PIB par habitant 2024
df['PIB_par_habitant_2024'] = df['PIB_2024_Md$'] * 1e9 / (df['Population_millions'] * 1e6)

# Calcul croissance PIB en % entre 2023 et 2024
df['Croissance_PIB_2024_%'] = ((df['PIB_2024_Md$'] - df['PIB_2023_Md$']) / df['PIB_2023_Md$']) * 100

# Trier par croissance décroissante 2024
df = df.sort_values(by='Croissance_PIB_2024_%', ascending=False)

# Génération rapport texte
with open("rapport_analyse_pib_2024.txt", "w", encoding='utf-8') as f:
    f.write("Rapport d'analyse du PIB mondial pour 2024\n\n")

    f.write(f"Nombre de pays analysés : {len(df)}\n\n")

    f.write("Classement des pays par croissance du PIB nominal en 2024 :\n")
    for i, row in df.iterrows():
        f.write(f"{row['Pays']}: PIB nominal 2024 = {row['PIB_2024_Md$']:.1f} Md$, croissance 2024 = {row['Croissance_PIB_2024_%']:.2f}%, ")
        f.write(f"PIB par habitant 2024 = {row['PIB_par_habitant_2024']:.0f} USD, prévision croissance 2025 = {row['Croissance_2025_%']}%\n")

    f.write("\nObservations générales :\n")
    f.write("- Les États-Unis et la Chine dominent toujours l'économie mondiale en termes de PIB nominal.\n")
    f.write("- L'Inde affiche la plus forte croissance parmi les grandes économies, avec une croissance attendue au-delà de 6%.\n")
    f.write("- Certains pays émergents montrent une amélioration significative, tandis que les économies européennes ont une croissance plus modérée.\n")
    f.write("- Les prévisions pour 2025 indiquent un maintien général de la croissance avec une légère variation selon les pays.\n")

print("Rapport texte généré : rapport_analyse_pib_2024.txt\n")

# Graphique croissance PIB 2024
plt.figure(figsize=(12, 8))
sns.barplot(x='Croissance_PIB_2024_%', y='Pays', data=df, palette='vlag')
plt.title('Croissance du PIB nominal des pays entre 2023 et 2024 (%)')
plt.xlabel('Croissance (%)')
plt.ylabel('Pays')
plt.tight_layout()
plt.show()

# Graphique PIB par habitant 2024
plt.figure(figsize=(12, 8))
sns.barplot(x='PIB_par_habitant_2024', y='Pays', data=df, palette='cool')
plt.title('PIB par habitant en 2024 (en dollars)')
plt.xlabel('PIB par habitant ($)')
plt.ylabel('Pays')
plt.tight_layout()
plt.show()

# Graphique prévision croissance 2025
plt.figure(figsize=(12, 8))
sns.barplot(x='Croissance_2025_%', y='Pays', data=df, palette='mako')
plt.title('Prévision de croissance du PIB pour 2025 (%)')
plt.xlabel('Croissance prévue (%)')
plt.ylabel('Pays')
plt.tight_layout()
plt.show()
```
